# Q_18__Access the last three elements of a list without using negative indices.


range in list___start:stop:step
2. Start from start
This means the list will begin at the index you write in place of start.
3. Go up to but not including stop
The slicing will stop before this index. This index itself is not included.

4. Move in jumps of step
Python will take every step-th element (like skipping).





```python
my_list=['abid','taha','aswad','ali','ahmed','musa']
#list[start:stop:step]

print(my_list[5:6])
```

```python
my_list=['abid','taha','aswad','ali','ahmed','musa']
#list[start:stop:step]

print(my_list[4:5])
```
